```python
# example program
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab as pl
#data preprocessing
iris = sns.load_dataset("iris") #加载公共的画图测试数据集，iris
data=iris[["sepal_length","sepal_width"]]# 获取sepal_length 跟 sepal_width列信息，并且只保存这两列的信息
data.fillna(0.1)# 填充 值为空（即为Nan， Na, nan的值 ）为0.1
# print data
# print data
# data=data.values
# print data
data=data[0:20]# 选取前 20维绘制图片，数据太多会影响美观
data.index=["1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3","1","3"] #可以设置index,很明显，index会显示在图的y轴上
#define a panel with size (12,12) and 300 dpi
plt.figure(figsize=(12,12), dpi=300,facecolor=(1, 1, 1))# 开始平铺画布

#adjust distrances of subplots
# plt.subplots_adjust(wspace =0.3, hspace =0.15)

#plot multi sub figures into one figure
ax=sns.heatmap(data,linewidths=.5,cmap="YlOrRd")  #绘制图片，data为数据，linewidth=.5代表线条宽度，cmap用来设定渐变色
ax.set_xlabel('')# 将x轴的标签取消
plt.yticks(Rotation=0) #设置y轴的旋转角度
plt.ylabel("heatmap_seaborn",fontsize=16) #设置y轴标签
#show figure 
plt.show()# 显示出来
# plt.savefig('test_boxplot_seaborn.pdf',bbox_inches='tight') #若想要保存就需要用这个，切记，用这个命令的时候就不能使用plt.show(),否则图片保存为空白
```
