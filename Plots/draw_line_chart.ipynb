```python
# used as plot program
import matplotlib.pyplot as plt
import numpy as np
import math
plt.figure(figsize=(12,4), dpi=300,facecolor=(1, 1, 1))#平铺画布，figsize用来设置长宽
# plt.tight_layout()#调整整体空白
plt.subplots_adjust(wspace =0.35, hspace =0.42)#调整子图间距


plt.subplot(121)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="iRNA-Methyl_f-score.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))#读取文件，进行文件处理，这一步需要自己实现
all_results
x=range(len(all_results))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)

#plt.annotata表示在图中写字，第一个参数就是显示的内容，第二个人参数xytext为位置
plt.annotate("["+str(max_value_index+1)+","+str(max_value)+"]",xytext=(max_value_index-0.2*length_x,max_value),xy=(max_value_index,max_value),fontsize=11)
plt.plot([max_value_index,max_value_index],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)#画直线
plt.plot(x,all_results)#画折线图
plt.title("f_classif (Dataset-S51)",fontsize=11)#设置剃头了
plt.legend(loc="upper right",fontsize=11)#设置legend的位置，这边表示右上

#设置坐标轴范围

plt.xlim([0, length_x]) #设置x轴坐标
plt.ylim((0.5, 1))#设置y轴坐标
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11) #设置y轴标签
plt.xlabel('Feature dimension',fontsize=11)#设置x轴标签
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks) #设置y轴刻度
my_x_ticks = np.arange(0, length_x, (length_x/5)/10*10)
plt.xticks(my_x_ticks)#设置x轴刻度
plt.text(-0.19*length_x*1.0,0.98,'(A)',fontsize='xx-large') #在文中写字


plt.subplot(122)

x=np.arange(1,10,0.01)
# plt.plot(x,np.sin(x))
file_location="iRNA-Methyl_mrmr.txt"
all_results=[]
with open(file_location,'r') as f:#默认打开模式就为r
    data=f.readlines()
    for yy in data:
        if yy.startswith("[[["):
            all_results.append(float(yy.strip()[3:-3].split(",")[1]))
all_results
x=range(len(all_results))
length_x=len(x)
max_value=max(all_results)
max_value_index=all_results.index(max_value)
max_value=round(max_value,4)
plt.annotate("["+str(max_value_index+1)+","+str(max_value)+"]",xytext=(max_value_index-0.2*length_x,max_value),xy=(max_value_index,max_value),fontsize=11)
plt.plot([max_value_index,max_value_index],[0,max_value],color="red",label="Best ACC: %.4f"%max_value)
plt.plot(x,all_results)
plt.title("mrmr (Dataset-S51)",fontsize=11)
plt.legend(loc="upper right",fontsize=11)


#设置坐标轴范围

plt.xlim([0, length_x])
plt.ylim((0.5, 1))
#设置坐标轴名称
plt.ylabel('Accuracy',fontsize=11)
plt.xlabel('Feature dimension',fontsize=11)
#设置坐标轴刻度
my_y_ticks = np.arange(0.5, 1, 0.1)
plt.yticks(my_y_ticks)
my_x_ticks = np.arange(0, length_x, (length_x/5)/10*10)
plt.xticks(my_x_ticks)
plt.text(-0.19*length_x*1.0,0.98,'(B)',fontsize='xx-large')

plt.show()#显示图片，不可与plt.savefig同时用

# plt.savefig('2RFH-feature_selection_chart.png',bbox_inches='tight') #保存图片，可以保存格式 pdf, svg,tiff。其中 bbox_inches用来是图片更加紧凑
# print "done"
```
